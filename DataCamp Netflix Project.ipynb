```python
# Importing the relevant libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Years list
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

# Movie lengths
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

# Creating a dictionary where years are keys, and durations as values
movie_dict = {'years':years , 'durations':durations}

# Print the dict
print(movie_dict)
```

```python
# Creating a dataframe from the dictionary 
durations_df = pd.DataFrame(movie_dict)

durations_df
```

```python
# Me a graph of the data 

x = durations_df['years']
y = durations_df['durations']
plt.plot(x, y)
plt.title("Netflix Movie Durations 2011-2020")

plt.show()
```

```python
# Read the csv file
netflix_df = pd.read_csv("Documents/netflix_data.csv")

# Print the first five elements
netflix_df.head()
```

```python
# Subset the dataframe
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

netflix_df_movies_only
```

```python
# Subset the columns

netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]

# Print first five rows
netflix_movies_col_subset[0:5]
```

```python
# Create a scatter plot

x = netflix_movies_col_subset['duration']
y = netflix_movies_col_subset['release_year']
plt.scatter(x, y)

# Create a title
plt.title("Movie Duration by Year of Release")

plt.show()
```

```python
# Subset netflix_movies_col_subset to create a new DataFrame short_movies containing only movies 
# that have a duration fewer than 60 minutes.

short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60 ]

# Print first 20 rows
short_movies.head(20)
```

```python
# Define an empty list colors
colors = []

# Iterate through the netflix_movies_col_subset DataFrame's rows and append colors to your colors
for lab, row in netflix_movies_col_subset.iterrows() :
    if row['genre'] == 'Children' :
        colors.append('red')
    elif row['genre'] == 'Documentaries' :
        colors.append('blue')
    elif row['genre'] == 'Stand-Up' :
        colors.append('green')
    else :
        colors.append('black')
        
# Print first ten values
colors[0:10]
```

```python
# Create a scatter plot of duration versus release_year
x = netflix_movies_col_subset["release_year"]
y = netflix_movies_col_subset["duration"]
plt.scatter(x, y, c = colors)

# Create a title and axis labels
plt.title("Movie duration by year of release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

plt.show()
```
